```python
import numpy as np
import cv2
from sklearn.model_selection import train_test_split
import pandas as pd
from PIL import Image 
import os
import glob
```

```python
data=pd.read_csv('/home/gharshi/Data_Annotation.csv', encoding='latin-1')
```

```python
data
```

```python
data["file_name"] = data['accident_no'] + "_" + data["img_name"]
```

```python
cdc=[col for col in list(data.columns) if "code" in col]
data_df=data.loc[pd.notnull(data[cdc]).all(axis=1),:]
data_df
```

```python
# shuffle = True
train,test=train_test_split(data_df.index, shuffle=True, test_size=0.2, random_state=1)
test,val=train_test_split(test, shuffle=True, test_size=0.5, random_state=2)
```

```python
def image1(folder,list_index):
    
    folder_path = '/home/gharshi/Severe_Accidents/'+folder+'/'
    try:
        os.mkdir(folder_path)
    except OSError:
        print ("Creation of the directory %s failed" % folder_path)

    
    for i in list_index:
        
        path = '/home/gharshi/Data/Affine/Filtered Images'
        img = data_df.loc[i,'database']+'/' + data_df.loc[i,'accident_no'] +'/' + data_df.loc[i,'img_name']
        try:
            img_1 = cv2.imread(glob.glob(path+'/'+img+'.*')[0])
            cv2.imwrite(folder_path+data_df.loc[i,'file_name'] + '.jpg', img_1) 
        except:
            print(img)
```

```python
image1('train',train)
image1('test',test)
image1('val',val)
```

```python
data_df.to_csv('/home/gharshi/Severe_Accidents/Data_Annotation.csv', encoding = 'latin-1')
```
